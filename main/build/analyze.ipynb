```python
# 我将根据您的要求修改代码，为每个指标创建单独的图表，并添加网格和数据点上的数值标签。

# 首先，我们更新DataFrame的列名，以匹配您提供的代码片段中使用的列名。
df.columns = ['Capacity', 'Height', 'InNN', 'LeafNN', 'ConstructionT', 'NNQueryT', 'RangeQueryT']

# 接下来，我们为每个指标创建单独的图表，并添加网格以及在数据点上标注数值。

def plot_metric_vs_capacity(metric_name, color, ylabel, title):
    plt.figure(figsize=(10, 4))
    plt.plot(df['Capacity'], df[metric_name], marker='o', color=color)
    plt.title(title)
    plt.xlabel('Capacity')
    plt.ylabel(ylabel)
    plt.grid(True)
    for x, y in zip(df['Capacity'], df[metric_name]):
        label = "{:.4f}".format(y)
        plt.annotate(label, (x, y), textcoords="offset points", xytext=(0, 10), ha='center')
    plt.show()

# 容量与树高度的关系图
plot_metric_vs_capacity('Height', 'blue', 'Height', 'Tree Height vs Capacity')

# 容量与内部节点数的关系图
plot_metric_vs_capacity('InNN', 'green', 'Interior Node Number', 'Interior Node Number vs Capacity')

# 容量与叶节点数的关系图
plot_metric_vs_capacity('LeafNN', 'red', 'Leaf Node Number', 'Leaf Node Number vs Capacity')

# 容量与构建时间的关系图
plot_metric_vs_capacity('ConstructionT', 'purple', 'Construction Time', 'Construction Time vs Capacity')

# 容量与最邻近查询时间的关系图
plot_metric_vs_capacity('NNQueryT', 'orange', 'NNQuery Time', 'NNQuery Time vs Capacity')

# 容量与范围查询时间的关系图
plot_metric_vs_capacity('RangeQueryT', 'brown', 'RangeQuery Time', 'RangeQuery Time vs Capacity')
```
